### Comparing across a few colleges

```python
def college_list(collegeid):
    return [i for i in range(len(df)) if df.OPEID[i] == collegeid]

# print college_list('217800') #mit
# print college_list('215500') #harvard

def colleges_list(collegeidlist):
    collegeslist = []
    for collegeid in collegeidlist:
        collegeslist.extend(college_list(collegeid))
    return collegeslist

collegeslist = colleges_list(['217800', '215500','142600','130500','3946300']) #mit, harvard, yale, stanford, olin
# collegeslist2 = colleges_list(['215500','142600','130500'])
dfcollege = pd.DataFrame()
for i in collegeslist:
    dfcollege = pd.concat([dfcollege, df[i:i+1]])
```

```python
print dmajor['PCIP14']
sns.pointplot(data=dfcollege, x = 'Year', y ='PCIP14', hue='OPEID')
```
